# 93. Hiyerarşik Kümeleme

* Hiyerarşik Kümeleme (SciPy)
* t-SNE : 2-boyutlu veri seti haritası/grafiği

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
wheat = data[:,:-1]
location = data[:,-1]
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
import matplotlib.pyplot as plt
```

```python
merge = linkage(wheat, 'complete')

fig = plt.figure(figsize = (35,30))
dn = dendrogram(merge, leaf_rotation = 90,
                leaf_font_size = 12, orientation = 'left')
plt.show()
```
