**Gözetimli Ve Gözetimsiz Öğrenme**

* Gözetimli Öğrenmede bir tahmin modeli geliştirilir.(Tahmin Var)
* Gözetimsiz Öğrenmede veri seti analiz edilir. (Tahmin Yok)

```python
from sklearn import datasets
iris = datasets.load_iris()
data = iris.data
target = iris.target
list(iris.target_names)
```

```python
print(data)
```

```python
# Kümeleme (Clustering)
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 3)
model.fit(data)
```

```python
kume = model.predict(data)
print(kume)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

x = data[:,0]
y = data[:,2]
plt.scatter(x, y, c = kume)
plt.xlabel('Sepal Lenght')
plt.ylabel('Petal Lenght')
plt.show()
```

```python
# Yeni Veriler
# Centorids
centroids = model.cluster_centers_
centroid_x = centroids[:,0]
centroid_y = centroids[:,2]
plt.scatter(centroid_x, centroid_y, marker = 'D', c = 'red', s = 10)
plt.show()
```

```python
plt.scatter(x, y, c = kume)
plt.scatter(centroid_x, centroid_y, marker = 'D', c = 'red', s = 100)
plt.xlabel('Sepal Lenght')
plt.ylabel('Petal Lenght')
plt.show()
```
