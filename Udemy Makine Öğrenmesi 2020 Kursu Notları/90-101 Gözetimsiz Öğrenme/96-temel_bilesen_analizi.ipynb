# 96. Temel Bileşen Analizi (Principal Component Analysis)

- Boyut azaltma (dimension reduction)
- Veri setindeki 'önemsiz' verileri elemine etmek
    * Bu tür veriler sınıflandırma, regresyon performansını azaltır.
- PCA = Principal Component Analysis
- En temel ve yaygın boyut azaltma yöntemidir.
    * Veriler 'de-korele' edilir.
        + Veriler üzerinde herhangi bir değişiklik yapılmaz.
        + Veriler eksenler doğrultusunda sıralanır. (Mean = 0)
        + Veri/Bilgi kaybı meydana gelmez.
    + Boyut azaltma

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')

lenght = data[:,3]#ilk değer satırları ifade eder : işareti tüm satırları alır.
width = data[:,4]
geometry = data[:,3:5]
```

```python
import matplotlib.pyplot as plt

plt.scatter(width, lenght)
plt.xlabel('Width of Kernel')
plt.ylabel('Lenght of Kernel')
plt.axis('equal')
plt.show()

from scipy.stats import pearsonr
correlation, pvalue = pearsonr(lenght, width)
print(correlation)
```

```python
from sklearn.decomposition import PCA
model = PCA()

pca_features = model.fit_transform(geometry)
x = pca_features[:,0]
y = pca_features[:,1]

plt.axis('equal')
plt.scatter(x, y)
plt.show()

correlation, pvalue = pearsonr(x, y)
print(correlation)
```
