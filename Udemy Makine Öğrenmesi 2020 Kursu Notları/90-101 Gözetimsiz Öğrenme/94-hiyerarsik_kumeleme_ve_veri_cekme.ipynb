# 94. Hiyerarşik Kümeleme Ve Veri Çekme

* Dendrogramlar sadece veri görüntülemesi için kullanılmazlar.
* Herhangi bir birleştirme aşamasında veri çekilebilir.
* Dendrogram yüksekliği = Birleştirilen kümeler arasındaki mesafe 
    + Verilen yükseklik değerinden sonra birleştirme işlemi yapılmaz.
    + fcluster()

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
wheat = data[:,:-1]
location = data[:,-1]

from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
merge = linkage(wheat, method = 'complete')
plt.figure(figsize = (20, 20))
dendrogram(merge, leaf_rotation = 90, 
           leaf_font_size = 12)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
import pandas as pd
```

```python
labels = fcluster(merge, 4, criterion = 'distance')
```

```python
labels
```

```python
df = pd.DataFrame({'Küme': labels, 'Yer' : location})
df['Yer'] = df['Yer'].replace({1 : ' Kama', 2 : 'Rosa', 3 : 'Canadian'})
crosstab = pd.crosstab(df['Küme'], df['Yer'])
print(crosstab)
```
