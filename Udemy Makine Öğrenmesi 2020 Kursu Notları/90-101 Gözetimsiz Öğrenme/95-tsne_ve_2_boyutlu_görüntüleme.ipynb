# 95. t-SNE ve 2 Boyutlu Görüntüleme

* t-distributed Stochastic Neighbor Embedding
* Verileri 2 boyutlu görüntülemeyi sağlar.
* Veri setlerini analiz etmek için kullanılır.
* KMeans, inertia, plot ile benzeri çıktılar elde edilir.
* fit_transform() metodu kullanılır.
    - fit() metodu yoktur.
    - transform() metodu yoktur.
    - Yeni verilere eklenemez.

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
wheat = data[:,:-1]
location = data[:,-1]
```

```python
from sklearn.manifold import TSNE
model = TSNE(learning_rate = 250)
tsne = model.fit_transform(wheat)
x = tsne[:,0]
y = tsne[:,1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
plt.scatter(x, y, c = location)
plt.show()
```
