# 63. Bee Swarm Plots 

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('2008_swing_states.csv')

plt.hist(data['dem_share'], bins = 10) #histogram oluşturma
plt.xlabel('Demokrat Oy Oranı')
plt.ylabel('İlçeler')
plt.title('Demokratların Swing State Oyları (2008 Seçimleri)')
plt.show()
```

```python
import seaborn as sns
sns.set()

sns.swarmplot(x='state',y='dem_share',data = data)
plt.xlabel('Eyalet')
plt.ylabel('Demokrat Oy Oranı')
plt.title('Demokratların Swing State Oyları (2008 Seçimleri)')
plt.show()
```
