# 67. Varyans ve Standart Sapma

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv('2008_swing_states.csv')
sns.set()
sns.swarmplot(x = 'state', y = 'dem_share', data = data)
plt.xlabel('Eyalet')
plt.ylabel('Demokrat Oy Oranı')
plt.title('Demokratların Swing States Oyları (2008 Seçimleri)')
plt.show()
```

```python
import numpy as np
data2 = data[data['state'] == 'FL']
data3 = data['dem_share']
```

```python
#Varyans hesaplama
np.var(data3) 
```

```python
# Standart sapma
np.std(data3)
```

```python
# Varyansın Karesi = Standart Sapma
np.sqrt(np.var(data3))
```
