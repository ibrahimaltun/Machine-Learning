# 66. YÃ¼zdelikler Ve Box Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

veri = pd.read_csv('2008_all_states.csv')
sns.set()

x = np.sort(veri['dem_share'])
y = np.arange(1, len(x) + 1) / len(x)

plt.plot(x,y, marker = '.', linestyle='none')
plt.xlabel('Oy OranÄ±')
plt.ylabel('ECDF')
plt.title('2008 SeÃ§imleri Oy Analizi')
plt.margins(0.05)
plt.show()
```

```python
# YÃ¼zdelik hesaplama
yuzdelik = np.percentile(veri['dem_share'], [10,25,50,75,98])
print(yuzdelik)
```

```python
# Box Plot oluÅturmak
sns.boxplot(x = 'east_west', y = 'dem_share', data = veri)
plt.xlabel('Eyalet (BatÄ±/DoÄu)')
plt.ylabel('AlÄ±nan Oy OranlarÄ±')
plt.title('2008 seÃ§im Analizi')
plt.rcParams['figure.figsize'] = [30,15]
plt.show()
```
