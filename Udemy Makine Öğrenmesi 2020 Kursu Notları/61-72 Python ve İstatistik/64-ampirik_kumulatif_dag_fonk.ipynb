## 64. Ampirik Kümülatif Dağılım Fonksiyonları(Empiricial Cumulative Distribution Function)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri = pd.read_csv('2008_swing_states.csv')
sns.set()
x = np.sort(veri['dem_share'])
y = np.arange(1, len(x)+1) / len(x)

plt.plot(x,y,marker= '.',linestyle='none')
plt.xlabel('Oy Oranı')
plt.ylabel('ECDP')
plt.title('2008 Seçimleri Oy Analizi')

plt.margins(0.05)


plt.show()
```

```python
x = np.sort(veri['rep_votes'])
y = np.arange(1, len(x)+1) / len(x)
plt.plot(x,y,marker= '.',linestyle='none')
plt.xlabel('Oy Oranı')
plt.ylabel('ECDP')
plt.title('2008 Seçimleri Oy Analizi')
plt.margins(0.05)

x1 = np.sort(veri['dem_votes'])
y1 = np.arange(1, len(x)+1) / len(x)
plt.plot(x1,y1,marker= '.',linestyle='none')
plt.xlabel('Oy Oranı')
plt.ylabel('ECDP')
plt.title('2008 Seçimleri Oy Analizi')
plt.margins(0.05)
plt.legend(('Cumhuriyetçiler','Demokratlar'))
plt.show()
```

```python
def secim(veri):
    """ECDF grafiği için x ve y değerlerini hesaplar"""
    n = len(veri) # toplam veri sayısını hesaplama
    x = np.sort(veri) # x eksenini tanımlama
    y = np.arange(1, n+1) / n 
    
    return x,y    
```

```python
x_dem, y_dem = secim(veri['dem_votes'])
x_rep, y_rep = secim(veri['rep_votes'])

plt.plot(x_dem, y_dem)
plt.plot(x_rep, y_rep)
plt.show()
```
