# 62. Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('2008_swing_states.csv')

plt.hist(data['dem_share']) #histogram oluşturma
plt.xlabel('Demokrat Oy Oranı')
plt.ylabel('İlçeler')
plt.title('Demokratların Swing State Oyları (2008 Seçimleri)')
plt.show
```

```python
plt.hist(data['dem_share'],bins = 75) #histogram oluşturma
plt.xlabel('Demokrat Oy Oranı')
plt.ylabel('İlçeler')
plt.title('Demokratların Swing State Oyları (2008 Seçimleri)')
plt.show()
```

```python
# Veri setinin uzunluğunu belirleme
bin_sayisi = len(data)
print(bin_sayisi)
bin_sayisi2 = bin_sayisi**0.5
print(bin_sayisi2)
```

```python
import numpy as np
bin_sayisi3 = np.sqrt(bin_sayisi)
print(bin_sayisi3)
```
