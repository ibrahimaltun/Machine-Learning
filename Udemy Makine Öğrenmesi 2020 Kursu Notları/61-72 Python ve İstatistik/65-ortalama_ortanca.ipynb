# 65. Ortalama ve Ortanca

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('2008_all_states.csv')

import seaborn as sns
sns.set()
data2 = data[data['state'] == 'UT']
sns.swarmplot(x = 'state', y = 'dem_share', data = data2)

plt.xlabel('Eyalet')
plt.ylabel('Demokrat Oy Oranları')
plt.title('Demokratların 2008 Seçimleri Oyları')
plt.show()
```
