# 74. Keşifsel Veri Analizi (Exploratory Data Analysis)

<b>İris Çiçeği Veri Seti(İris Dataset)</b>

<b>Features (Özellikler)</b>

- Taç yaprak uzunluğu (Petal length)
- Taç yaprak genişliği (Petal width)
- Çanak yaprak uzunluğu (Sepal length)
- Çanak yaprak genişliği (Sepal width)

<b>Target Variable (Hedef Değişkenler)</b>

        o Versicolor
        o Verginica
        o Setosa

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
iris = datasets.load_iris()
```

```python
print(type(iris))
```

```python
print(iris.keys())
```

```python
iris.target_names
```

```python
iris.target_names.shape
```

```python
# DataFrame'e çevirme (Veri çerçevesi)
x = iris.data
y = iris.target
veri = pd.DataFrame(x, columns = iris.feature_names)
print(veri.head())
```

```python
# Grafiksel Keşifsel Veri Analizi
pd.plotting.scatter_matrix(veri, c = y, figsize = [10,10], s = 50)
plt.show()
```
