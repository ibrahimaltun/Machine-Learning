# 77. Regresyon - Regression

Hedef değişken (target variable) bir sabit değil aksine değişkendir

    - Bir evin fiyatı
    - Bir ülkenin toplam gayri safi milli hasılası
    - Bir işletmenin aylık geliri

```python
import pandas as pd
data = pd.read_csv('boston.csv')
```

```python
data.head()
```

```python
# sci-kit learn 
# RM ve MEDV değerleri ayrı ayrı saklanmalıdır.

X = data.drop('medv', axis = 1).values #axis =0: satırlar, axis= 1 : sütunlar
y = data['medv'].values #sadece target variable
```

```python
# Tek bir parametre kullanmak
Room_Number = X[:,6]
Room_Number = Room_Number.reshape(-1,1) # -1 : Tüm satırları seçtiriyor.
y = y.reshape(-1,1)
```

```python
# Grafik oluşturma
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.scatter(Room_Number,y)
plt.title('Oda sayısı ile ev fiyatı arasındaki ilişki')
plt.xlabel('Oda Sayısı')
plt.ylabel('Ev Fiyatı (x1000$)')
plt.show()
```

```python
# Regresyon Modeli Oluşturma
import numpy as np
from sklearn.linear_model import LinearRegression

reg  = LinearRegression()
reg.fit(Room_Number, y)

regression_line = np.linspace(min(Room_Number), max(Room_Number))

plt.scatter(Room_Number, y)
plt.title('Oda sayısı ile ev fiyatı arasındaki ilişki')
plt.xlabel('Oda Sayısı')
plt.ylabel('Ev Fiyatı (x1000$)')
plt.plot(regression_line, reg.predict(regression_line), color = 'yellow', linewidth = 3)
plt.show()
```
