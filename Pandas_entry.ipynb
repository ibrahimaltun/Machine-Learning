- Yalnızca istenilen aralıkta ki değerleri gösterme işlemi.

```python
above_35 = titanic['Age'] > 35 # Bu şekilde sadece true false döndürür.
above_35 
```

```python
above_35 = titanic[titanic['Age'] > 35] #Böyle ise değerleri döndürür.
above_35 
```

- Sadece 2. ve 3. kabinde ki yolcuları görmek istersek

```python
class_23 = titanic[titanic['Pclass'].isin([2,3])]
class_23
```

- Eğer istenilen bir veriye ait bir değer istiyorsak loc kullanırız.

```python
adult_names = titanic.loc[titanic['Age'] > 35, 'Name']
# Yani 35 yaşından büyük kişilerin isimlerini göster.
adult_names
```

- Eğer belirli bir sütun ve satırdaki verileri istersek iloc komutunu kullanırız.

```python
titanic.iloc[9:25, 2:5] # Satırda 9 dahil 25 değil ve sütunda 2 dahil 5 değil.
```
